```python
import json
import pandas as pd
```

```python
# set pandas to display all the columns
pd.set_option('display.max_columns', None)
```

```python
# load json file
with open('weather_data.json') as json_file:
    data = json.load(json_file)
```

```python
# Extract the 'data' part from the json file and transform it into a DataFrame
df = pd.DataFrame(data['data'])

# display top 10 rows of dataframe
df.head(10)
```

```python
# Convert the DataFrame to a list of dictionaries
data = df.to_dict(orient='records')

cleaned_data = []

# Iterate over each entry
for entry in data:
    # Extract the desired columns
    cleaned_entry = {
        'temp': entry['temp'],
        'datetime': entry['datetime'],
        'weather': entry['weather']
    }
    cleaned_data.append(cleaned_entry)

# Write the cleaned data to a new JSON file
with open('output.json', 'w') as file:
    json.dump(cleaned_data, file)
```

```python
# Read the JSON file
with open('output.json') as file:
    data = json.load(file)

# Create a DataFrame from the JSON data
df = pd.DataFrame(data)

# Display the top 10 rows of the DataFrame
df.head(10)
```

```python
df['weather_description'] = df['weather'].apply(lambda x: x['description'])

# Display the updated DataFrame
df.head(10)
```

```python
# Assuming you have the 'weather_description' column in the DataFrame
unique_values = df['weather_description'].nunique()

print("Number of unique values in 'weather_description' column:", unique_values)
```

```python
# Assuming you have the 'weather_description' column in the DataFrame
unique_values = df['weather_description'].unique().tolist()

# Group the unique values based on specific categories
snow_group = [value for value in unique_values if 'snow' in value.lower()]
flurries_group = [value for value in unique_values if 'flurries' in value.lower()]
haze_group = [value for value in unique_values if 'haze' in value.lower()]
clouds_group = [value for value in unique_values if 'clouds' in value.lower()]

print("Grouping of unique values in 'weather_description' column:")
print("Snow Group:", snow_group)
print("Flurries Group:", flurries_group)
print("Haze Group:", haze_group)
print("Clouds Group:", clouds_group)
```

```python
# Group the unique values based on specific categories
clouds_group = [value for value in unique_values if 'clouds' in value.lower() or 'haze' in value.lower()]
snow_group = [value for value in unique_values if 'snow' in value.lower()]
flurries_group = [value for value in unique_values if 'flurries' in value.lower()]

print("Grouping of unique values in 'weather_description' column:")
print("Clouds Group:", clouds_group)
print("Snow Group:", snow_group)
print("Flurries Group:", flurries_group)
```

```python

# Group the unique values based on specific categories
clouds_group = [value for value in unique_values if 'clouds' in value.lower() or 'haze' in value.lower()]
snow_group = [value for value in unique_values if 'snow' in value.lower() or ('flurries' in value.lower() and 'mix snow/rain' not in value.lower())]
print("Grouping of unique values in 'weather_description' column:")
print("Clouds Group:", clouds_group)
print("Snow Group (including Flurries):", snow_group)
```

```python
# Update the 'weather_description' column with the desired categories
df.loc[df['weather_description'].str.contains('snow', case=False), 'weather_description'] = 'Snow'
df.loc[df['weather_description'].str.contains('flurries', case=False), 'weather_description'] = 'Snow'
df.loc[df['weather_description'].str.contains('clouds|haze', case=False), 'weather_description'] = 'Cloudy'

# Verify the updated 'weather_description' column
print(df['weather_description'].unique().tolist())
```
